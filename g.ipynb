We start by defining our function

```python
def f(x):
    return x**2
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
```

<font color="orange">We create a set of x-values we wish to compute</font>

|x|y|
|-|-|
|0|0|
|1|1|
|2|4|

etc ...

```python
y = f(x)
```

```python
from matplotlib import pyplot as plt
```

Now display each plot as different figures

```python
plt.figure(1)
plt.plot(x, y)

plt.figure(2)
plt.plot(-x, y)

plt.figure(3)
plt.plot(x, -y)

plt.figure(4)
plt.plot(-x, -y)

plt.show()
```

We can also combine the figures into subplots

```python
fig, axis = plt.subplots(2, 2)

# Fixes titles overlapping axis labels
plt.tight_layout()

axis[0, 0].plot(x, y)
axis[0, 0].set_title('x^2')

axis[0, 1].plot(-x, y)
axis[0, 1].set_title('(-x)^2')

axis[1, 0].plot(x, -y)
axis[1, 0].set_title('-x^2')

axis[1, 1].plot(-x, -y)
axis[1, 1].set_title('-(-x)^2')

plt.show()
```

Lastly, we can plot everything on the same plot

```python

plt.plot(x, y, label='x^2')
plt.plot(-x, y, label= '(-x)^2')
plt.plot(x, -y, label= '-x^2')
plt.plot(-x, -y, label= '-(-x)^2')

# Show the labels
plt.legend()

plt.show()
```
